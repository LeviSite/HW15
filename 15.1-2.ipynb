```python
import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv("BRCA_pam50.tsv", sep="\t", index_col=0)
X = df.iloc[:, :-1].to_numpy()
y = df["Subtype"].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=17)

model = KNeighborsClassifier(n_neighbors=1, p=2)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(accuracy_score(y_pred, y_test))
print(y_pred)
print(y_test) #true values

#problem 2
matches = {"Luminal A":0, "Healthy":0, "Luminal B":0, "Triple-negative":0, "HER2-enriched":0, "Normal-like":0}
total = {"Luminal A":0, "Healthy":0, "Luminal B":0, "Triple-negative":0, "HER2-enriched":0, "Normal-like":0}
for i in range(len(y_test)):
    total[y_test[i]]+=1
    if y_pred[i] == y_test[i]:
        matches[y_test[i]]+=1
accuracy = {"Luminal A":matches["Luminal A"]/total["Luminal A"], "Healthy":matches["Healthy"]/total["Healthy"], 
            "Luminal B":matches["Luminal B"]/total["Luminal B"], "Triple-negative":matches["Triple-negative"]/total["Triple-negative"], 
            "HER2-enriched":matches["HER2-enriched"]/total["HER2-enriched"], "Normal-like":matches["Normal-like"]/total["Normal-like"]}
print(accuracy)
```
